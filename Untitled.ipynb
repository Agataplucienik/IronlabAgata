```python
import pandas as pd
from bs4 import BeautifulSoup
import pprint
import requests
import random
import re
import numpy as np
```

```python
url="https://www.cleveland.com/entertainment/2022/03/the-100-greatest-pop-punk-songs-of-all-time.html"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
soup.select("b")
```

```python
total=[]
for p in soup.select("b"):
    total.append((p.get_text().strip()))
```

```python
split_=[]
for t in total:
    split_.append(t.split("-"))
```

```python
df=pd.DataFrame(total)
```
