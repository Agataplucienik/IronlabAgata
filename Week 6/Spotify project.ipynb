```python
import config
import pprint
```

```python
import spotipy
import pandas as pd
import json
from spotipy.oauth2 import SpotifyClientCredentials


#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= config.client_id,
                                                           client_secret= config.client_secret))
```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
#germany
```

```python
germany =get_playlist_tracks('spotify','37i9dQZEVXbJiZcmkrIHGU')
```

```python
germany
```

```python
germany_id_list = []
for i in germany:
    germany_id_list.append(i['track']['id'])
```

```python
germany_list =[]
for i in germany_id_list:
    germany_list.append(sp.audio_features(i)[0])
germany_list
```

```python
germany_df = pd.DataFrame(germany_list)
germany_df
```

```python
germany_df.to_csv("germany.csv")
```
