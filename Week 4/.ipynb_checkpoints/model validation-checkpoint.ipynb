```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\agata\Downloads\marketing_customer_analysis_clean.csv")
```

```python
df.head()
```

```python
df=df.select_dtypes(include="number")
```

```python
y=df[["total_claim_amount"]]
```

```python
y
```

```python
X=df.drop("total_claim_amount", axis=1)
```

```python
X
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale= StandardScaler()
```

```python
scale.fit(X_train)
```

```python
transformed_train_X=scale.transform(X_train)
```

```python
transformed_train_X
```

```python
lm=LinearRegression()
```

```python
model=lm.fit(transformed_train_X, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
transformed_test_X=scale.transform(X_test)
```

```python
y_pred=model.predict(transformed_test_X)
```

```python
y_test
```

```python
y_pred[:,0]
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
mse(y_test, y_pred)**0.5
```

```python
real_vs_pred = pd.DataFrame({'y_test': y_test["total_claim_amount"], 'y_pred':y_pred[:,0]})
```

```python
real_vs_pred
```

```python
from sklearn.metrics import mean_absolute_error as mae
```

```python
mae(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test, y_pred)
```

```python
r2
```

```python
Adj_r2 = 1-(((1-r2)*(7637-1))/(7637-9-1))
Adj_r2
```

```python
import statsmodels.api as sm

X_train_const = sm.add_constant(X_train.to_numpy())

mode_l = sm.OLS(y_train, X_train_const).fit()
predictions_train = mode_l.predict(X_train_const)

X_test_const = sm.add_constant(X_test)
y_pred = mode_l.predict(X_test_const)
print_model = mode_l.summary()
print_model
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,4))
ax[0].plot(y_pred, y_test['total_claim_amount'], 'o')
ax[0].set_xlabel("y_test")
ax[0].set_ylabel("y_pred")
ax[0].set_title("Test Set -Predicted vs real")

# Get a histogram of the residuals ie: y - y_pred.
# Does it ressemble a normal distribution?
ax[1].hist(y_test['total_claim_amount'] - y_pred)
ax[1].set_xlabel("Test y-y_pred")
ax[1].set_title("Test Set Residual histogram")

# Homoscedasticity
ax[2].plot(y_pred,np.array(y_pred)-np.array(y_test['total_claim_amount']),"o")
ax[2].set_xlabel("predicted")
ax[2].set_ylabel("residuals")
ax[2].set_title("Residuals by Predicted")
ax[2].plot(y_pred,np.zeros(len(y_pred)),linestyle='dashed')
```

```python
X.columns
```

```python
model.coef_[0,:]
```

```python
feature_importances
```
