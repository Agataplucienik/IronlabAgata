```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\agata\Downloads\marketing_customer_analysis_clean.csv")
```

```python
df=df.select_dtypes(include="number")
```

```python
y=df[["total_claim_amount"]]
```

```python
y
```

```python
X=df.drop("total_claim_amount", axis=1)
```

```python
X
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale= StandardScaler()
```

```python
scale.fit(X_train)
```

```python
transformed_train_X=scale.transform(X_train)
```

```python
lm=LinearRegression()
```

```python
model=lm.fit(transformed_train_X, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
transformed_test_X=scale.transform(X_test)
```

```python
y_pred=model.predict(transformed_test_X)
```

```python
y_test
```

```python
y_pred[:,0]
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
mse(y_test, y_pred)**0.5
```
